## Gradient Boost Regressor for conversion

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.grid_search import GridSearchCV
```

```python
y = df[(~df['conversion_rate'].isna()) & (~df['ad_copy'].isna())]['conversion_rate']
X_train, X_test, y_train, y_test = train_test_split(tf_mat, y, test_size=.3)
```

```python
gb_grid_params = {
              'n_estimators': [20,200,60],
              'learning_rate':[0.02, 0.01],
              'max_depth': [4, 6, 8],
              'min_samples_leaf': [20, 50,100,150],
              'max_features': [.5, 0.3, 0.1] 
              }
gridsearch = GridSearchCV(GradientBoostingRegressor(), gb_grid_params, n_jobs=-1,verbose=3)
```

```python
gridsearch.fit(X_train, y_train)
```

```python
best = gridsearch.best_estimator_
best
```

```python
best.score(X_test, y_test)
```

## Trying Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
param_grid = { 
            "n_estimators"      : [10,20,30,60,100],
            "max_features"      : ["auto", "sqrt", "log2"],
            "min_samples_split" : [2,4,8],
            "bootstrap": [True, False],
            }
random_grid = GridSearchCV(RandomForestRegressor(), param_grid, n_jobs=-1)
```

```python
random_grid.fit(X_train, y_train)
```

```python
best_rand = random_grid.best_estimator_
best_rand.score(X_test, y_test)
```
