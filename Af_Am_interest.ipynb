## Gradient Boost regressor for conversion rate

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.grid_search import GridSearchCV
```

```python
y = af[(~af['conversion_rate'].isna()) & (~af['ad_copy'].isna())]['conversion_rate']
X_train, X_test, y_train, y_test = train_test_split(count_mat, y, test_size=.3)
```

## Grid Search

```python
gb_grid_params = {
              'n_estimators': [60,100,150],
              'learning_rate':[0.02, 0.01],
              'max_depth': [4, 6, 8],
              'min_samples_leaf': [20, 50,100,150],
              'max_features': [.5, 0.3, 0.1] 
              }
gridsearch = GridSearchCV(GradientBoostingRegressor(), gb_grid_params, n_jobs=-1,verbose=3)
```

```python
gridsearch.fit(X_train, y_train)
```

```python
best = gridsearch.best_estimator_
```

```python
best
```

```python
best.score(X_test, y_test)
```

## Out of the box classifier

```python
grad = GradientBoostingRegressor(learning_rate=.1)
grad.fit(X_train, y_train)
```

```python
grad.score(X_test, y_test)
```
